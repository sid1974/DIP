```python
from PIL import Image
from IPython.display import display

# Load the image
image = Image.open("CAR.jpg")

# Get the size of the image
width, height = image.size

# Define quadrant labels
labels = ["Up Left", "Up Right", "Down Left", "Down Right"]

# Split the image into 4 quadrants
quadrants = [
    (labels[0], image.crop((0, 0, width//2, height//2))),   # up left
    (labels[1], image.crop((width//2, 0, width, height//2))),   # up right
    (labels[2], image.crop((0, height//2, width//2, height))),   # down left
    (labels[3], image.crop((width//2, height//2, width, height))),   # down right
]

# Display each quadrant with label
for label, quadrant in quadrants:
    print(label)
    display(quadrant)
```
